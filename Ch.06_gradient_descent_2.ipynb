```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
target = torch.FloatTensor([round((i+1)*0.1, 1) for i in range(9)]).reshape(3, -1)
print(target)
print(target.shape)
x = torch.rand_like(target)
x.requires_grad = True
print(x)
print(x.shape)
```

```python
mse_loss = nn.MSELoss()
loss = mse_loss(x, target)
print(loss)
```

```python
threshold = 1e-5
learning_rate = 0.1
iter_cnt = 0

while loss > threshold:
    iter_cnt += 1
    if x.grad is not None:
        x.grad.zero_()

    loss.backward()

    with torch.no_grad():
        x = x - (learning_rate * x.grad)

    x.detach_()
    x.requires_grad_(True)

    loss = mse_loss(x, target)

    print(iter_cnt, loss)
    print(x)
```
